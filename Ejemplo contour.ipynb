```python
from numpy import *
from matplotlib.pyplot import *
```

```python
n=40 # n el numero de nodos
maxi=15 # maxi el valor maximo
x=linspace(-maxi,maxi,n)
y=linspace(-maxi,maxi,n)
X,Y = meshgrid(x,y)
```

Para este ejemplo usaremos el potencial de una carga puntual en el origen
\begin{equation}
\phi=\frac{1}{4\pi \epsilon_0}\frac{q}{\sqrt{x^2 + y^2}}
\end{equation}

```python
Phi=1./sqrt(X**2 + Y**2)
```

```python
contour(X,Y,Phi,[0.1,0.2,0.3],colors=('b', 'g','r'))
```

Es algo complicado agregarle nombre a los contornos pero aquí les dejo un ejemplo

```python
co=contour(X,Y,Phi,[0.1,0.2,0.3],colors=('b', 'g','r'))
xlim(-12,12)
ylim(-12,12)
###########################
fmt={}
strs =[u'Electrodinámica','Hola','$\phi=0.3$']
for p,s in zip( co.levels, strs ):
    fmt[p] = s
clabel(co,co.levels,inline=True,fmt=fmt,fontsize=12)
###########################
```
